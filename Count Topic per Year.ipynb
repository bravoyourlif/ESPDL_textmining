```python
'''
 -    
'''

import csv
import pandas as pd


topicNum = 0
year = 2000

yearList = dict()
for i in range(2000,2020):
        yearList[i] = 0


while topicNum<18:
    inputFile = open('Article(Global)/global_final_beforeoount.csv','r')
    rdr = csv.reader(inputFile)
    for line in rdr:
        if int(line[2]) == topicNum:
            yearList[int(line[1])] += 1
    print("topic: ",topicNum)
    print(yearList)  
    df = pd.DataFrame(list(yearList.items()),columns = ['year','article']) 
    display(df)
    for i in range(2000,2020):
        yearList[i] = 0
    topicNum += 1

```
