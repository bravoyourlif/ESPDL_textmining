```python
'''
연도별 논문 갯수 출력
'''

import csv
import pandas as pd


topicNum = 0
year = 2000

yearList = dict()
for i in range(2000,2020):
        yearList[i] = 0

inputFile = open('Article(Korean)/bangje2.csv','r')
rdr = csv.reader(inputFile)
lineNum = 1

for line in rdr:
    if lineNum !=1 and line[1]!='':
        yearList[int(line[0])] += 1
    lineNum += 1

print(yearList)  
```

```python
df = pd.DataFrame(list(yearList.items()),columns = ['year','article']) 

display(df)
```
